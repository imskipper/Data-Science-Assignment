```python
import pandas as pd
```

```python
#Task 1 : Load the patient dataset and show summary with info().

patient = pd.read_csv('/content/Patient_Data.csv')
bill = pd.read_csv('/content/Billing_Data.csv')
print(patient)
print()
print(patient.info())
print()
print(bill)
print()
print(bill.info())
```

```python
#Task 2 : Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].

Billing_cols = patient.drop(['Name', 'ReceptionistID', 'CheckInTime'], axis=1)
print(Billing_cols)
```

```python
# Task 3 :	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

patient_cleaned = patient.drop(columns=['ReceptionistID', 'CheckInTime'])
print(patient_cleaned)
```

```python
#Task 4 : Use groupby to find total bill amount per department.

Total_bill_amount_per_department = patient_cleaned.groupby('Department')['BillAmount'].sum()
print(Total_bill_amount_per_department)
```

```python
#Task 5 :	Remove duplicate patient records based on PatientID.

patient = patient.drop_duplicates('PatientID')
print(patient)
```

```python
#Task 6 : Fill missing BillAmount values with the mean bill amount.
# find id there any null value
print(patient.isnull().sum())
print(patient)

# find MEAN
Mean = patient['BillAmount'].mean()
print("Mean is :", Mean)

#FILLNA
patient.fillna(Mean, inplace=True)
print(patient)
```

```python
# Task 7 :	Merge the billing dataset with patient dataset on PatientID.
Patient_and_Bill = pd.merge(patient, bill, on='PatientID')
print(Patient_and_Bill)
```

```python
#Task 8 : Concatenate an additional DataFrame that contains new patients for the current week (row-wise).
new_patients = pd.DataFrame({
    'PatientID': [106, 107],
    'Name': ['Fiona', 'George'],
    'Department': ['Neurology', 'Orthopedics'],
    'Doctor': ['Dr. John', 'Dr. Lee'],
    'BillAmount': [5800.0, 7200.0],
    'ReceptionistID': [1, 2],
    'CheckInTime': ['2023-08-01 11:45', '2023-08-01 12:45'],
    'InsuranceCovered': ['2500','1600'],
    'FinalAmount': ['4300','5600']
    })

df_extended = pd.concat([Patient_and_Bill, new_patients], ignore_index=True)
print(df_extended)
```

```python
#Task 9 : Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

# Calculate and add DiscountedAmount directly as a new column to df_extended
df_extended['DiscountedAmount'] = df_extended['BillAmount'] * 0.9

df_final = df_extended.copy()

print(df_final)
```
