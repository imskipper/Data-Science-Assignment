```python
# Question 1
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number to check if it's prime: "))
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
# Question 2
import random

# two random integers between 1 and 10
a = random.randint(1, 10)
b = random.randint(1, 10)

print(f"What is the product of {a} and {b}?")
user_answer = int(input("Your answer: "))

if user_answer == a * b:
    print("✅ Correct!")
else:
    print(f"❌ Incorrect. The correct answer is {a * b}.")
```

```python
# Question 3.1 :Printing squares of EVEN numbers between 100 and 200
print("Squares of even numbers from 100 to 200:")
for num in range(100, 201):
    if num % 2 == 0:
        print(f"{num}² = {num**2}")
```

```python
# Question 3.2 :Printing squares of ODD numbers between 100 and 200
print("Squares of odd numbers from 100 to 200:")
for num in range(100, 200):
    if num % 2 == 1:
        print(f"{num}² = {num**2}")
```
