```python
# Handle missing values
# In the Adult dataset, missing values are often represented as ' ?'
# We replace these with NaN to make them easier to handle
df.replace(' ?', np.nan, inplace=True)
df.replace('?', np.nan, inplace=True)
```

```python
# Define numerical features for scaling
numerical_features = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
```

```python
# Apply Scaling Techniques

# Standard Scaling (Z-score Normalization)
# Centers data around 0 with a standard deviation of 1
scaler_standard = StandardScaler()
df_standard = df.copy()
df_standard[numerical_features] = scaler_standard.fit_transform(df[numerical_features])

#  Min-Max Scaling (Normalization)
# Scales data to a fixed range [0, 1]
scaler_minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[numerical_features] = scaler_minmax.fit_transform(df[numerical_features])

#  Save the results
df_standard.to_csv('adult_standard_scaled.csv', index=False)
df_minmax.to_csv('adult_minmax_scaled.csv', index=False)

# Optional: Verify the scaling results
print("--- Standard Scaled Numerical Stats ---")
print(df_standard[numerical_features].describe().loc[['mean', 'std', 'min', 'max']])

print("\n--- Min-Max Scaled Numerical Stats ---")
print(df_minmax[numerical_features].describe().loc[['mean', 'std', 'min', 'max']])
```

## A. Standard Scaling (Z-score Normalization)
###How it works: It transforms data to have a mean of $0$ and a standard deviation of $1$.

###*When to prefer it:*

###Gaussian Distribution: When your features follow a Normal (Gaussian) distribution.

###Algorithm Assumptions: It is preferred for algorithms that assume the data is centered around zero or follow a normal distribution, such as Linear Regression, Logistic Regression, and Linear Discriminant Analysis (LDA).

###Distance/Variance Based: Essential for Principal Component Analysis (PCA) and Support Vector Machines (SVM) with RBF kernels.

###Outliers: It is generally more robust to outliers than Min-Max scaling, as it doesn't compress the majority of data into a tiny range due to a single extreme value.

##B. Min-Max Scaling (Normalization)

###How it works: It scales the data to a fixed range, usually $[0, 1]$, by subtracting the minimum and dividing by the range.

###*When to prefer it:*

###No Distribution Assumption: When you do not know the distribution of your data or if it is definitely not Gaussian.

###Neural Networks: Often used in Deep Learning as it helps gradients converge faster when inputs are in a small, bounded range.

###Magnitude Sensitivity: Useful for algorithms that are sensitive to the absolute scale of features, such as k-Nearest Neighbors (k-NN) and Recommender Systems.

###Image Processing: Standard practice for pixel intensities (scaling 0–255 to 0–1).Constraint: Should be avoided if there are significant outliers, as they will "squash" the normal data points into a very small interval (e.g., all values become 0.001 because one outlier is at 1.0).

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv('/content/adult_with_headers (1).csv')

# Basic Data Exploration
print("--- Data Info ---")
print(df.info())

print("\n--- Summary Statistics ---")
print(df.describe())

print("\n--- Missing Values Check ---")
# Replacing ' ?' with NaN for accurate counting
df.replace(' ?', np.nan, inplace=True)
df.replace('?', np.nan, inplace=True)
print(df.isnull().sum())

# Handle Missing Values
# Impute categorical columns with the mode (most frequent value)
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Encoding Techniques based on Category Count
unique_counts = {col: df[col].nunique() for col in categorical_cols}

# Lists to store columns for each encoding type
ohe_cols = [col for col in categorical_cols if unique_counts[col] < 5]
label_cols = [col for col in categorical_cols if unique_counts[col] >= 5]

print("\nOne-Hot Encoding applied to:", ohe_cols)
print("Label Encoding applied to:", label_cols)
```

```python
# Apply One-Hot Encoding (OHE)
# drop_first=True is often used to avoid the dummy variable trap
df_encoded = pd.get_dummies(df, columns=ohe_cols, drop_first=True)

# Apply Label Encoding
le = LabelEncoder()
for col in label_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Review final processed data
print("\n--- Final Data Types ---")
print(df_encoded.dtypes)
print("\n--- Processed Data Head ---")
print(df_encoded.head())

# Save the encoded dataframe
df_encoded.to_csv('adult_encoded.csv', index=False)
```

#Discussion: Pros and Cons of Encoding Techniques

#**One-Hot Encoding**

##Pros:

No Ordinal Bias: It treats all categories as equal, which is ideal for nominal data (where there is no inherent ranking).

Distance Friendly: Works well for algorithms like k-Nearest Neighbors or SVMs that rely on calculating distances between points.

##Cons:

Dimensionality Issues: Can create a massive number of features if the category count is high, leading to increased memory usage and training time.

Multicollinearity: Can introduce redundancy (often solved by using drop_first=True).

#**Label Encoding**

##Pros:

Memory Efficient: Keeps the dataset compact by using a single column of integers.

Tree-Model Compatible: Decision trees and Random Forests can naturally handle integer labels by splitting the data at specific values.

##Cons:

Artificial Ranking: Linear models may misinterpret the integers as a ranked order (e.g., thinking a category assigned '4' is "greater than" a category assigned '1').

Not Ideal for Distance-Based Models: Can lead to poor results in models like k-Means clustering or Linear Regression because the numerical "distance" between labels is meaningless.

```python
import pandas as pd
import numpy as np

# 1. Load and handle basic missing values first
df = pd.read_csv('/content/adult_with_headers (1).csv')
df.replace(' ?', np.nan, inplace=True)
df.replace('?', np.nan, inplace=True)
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# 2. Create New Features
# Feature 1: Net Capital Movement
df['net_capital'] = df['capital_gain'] - df['capital_loss']

# Feature 2: Work-Education Interaction
df['work_edu_index'] = df['hours_per_week'] * df['education_num']

print("--- New Features Created ---")
print(df[['net_capital', 'work_edu_index']].head())

# 3. Transformation for Skewed Features
# Checking skewness before transformation
original_skew = df['capital_gain'].skew()
print(f"\nSkewness of 'capital_gain' before transformation: {original_skew:.2f}")

# Applying Log Transformation (using log1p to handle 0 values safely)
df['capital_gain_log'] = np.log1p(df['capital_gain'])

# Checking skewness after transformation
new_skew = df['capital_gain_log'].skew()
print(f"Skewness of 'capital_gain' after Log transformation: {new_skew:.2f}")

# Save the updated dataset
df.to_csv('adult_feature_engineered.csv', index=False)
```

#Feature Engineering & Transformations
##New Features
##*Creatednet_capital:* Calculated as capital_gain - capital_loss.
* Rationale: Instead of treating gains and losses as independent events, the net value represents the actual financial growth or deficit for an individual. This provides a more holistic view of a person's financial status in a single feature.

##*work_edu_index:* Calculated as hours_per_week * education_num.

* Rationale: High income is typically a function of both the quantity of work (hours) and the quality/level of skill (education). This interaction feature captures the combined effect of labor and intellectual capital.


##Numerical Transformation
##**Feature Selected:**
capital_gain
##**Transformation Applied:**
Log Transformation (log1p)
##**Rationale:**
 The capital_gain feature is highly right-skewed (Skewness $\approx 11.95$). This indicates that most people have zero capital gains, while a few have very high values. Such skewness can negatively impact model performance, especially for linear models. Applying log1p ($log(1+x)$) compresses the range and reduces the skewness (reducing it to $\approx 3.09$), making the distribution more manageable for the algorithm.
