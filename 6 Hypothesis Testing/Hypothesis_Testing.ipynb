## SECTION 1: HYPOTHESIS STATEMENTS

```python
# ---------------------------------------------------------
# Step 1: State the Hypotheses
# ---------------------------------------------------------
print("="*60)
print("STEP 1: HYPOTHESIS STATEMENTS")
print("="*60)
print("\nNull Hypothesis (H₀): μ = Rs. 4,000")
print("   The weekly operating costs are equal to the theoretical mean.")
print("\nAlternative Hypothesis (H₁): μ > Rs. 4,000")
print("   The weekly operating costs are HIGHER than the theoretical mean.")
print("\nTest Type: One-tailed (right-tailed) test")
print("   We use a right-tailed test because owners claim costs are HIGHER.\n")
```


## SECTION 2: DATA INITIALIZATION

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# Step 2: Initialize Data and Calculate Parameters
# ---------------------------------------------------------
sample_mean = 3050          # x̄ = Sample mean weekly cost
units_mean = 600            # μ_X = Mean of X (units produced)
units_std = 25              # σ_X = Standard deviation of X
n = 25                      # Sample size
alpha = 0.05                # Significance level (5%)
```

## SECTION 3: THEORETICAL MEAN CALCULATION

```python
# Calculate theoretical mean using the cost model: W = 1000 + 5X
theoretical_mean = 1000 + 5 * units_mean  # μ = 1000 + 5(600) = 4000
theoretical_sigma = 5 * units_std         # σ = 5 × 25 = 125

print("="*60)
print("STEP 2: CALCULATE TEST STATISTIC")
print("="*60)
print(f"\nGiven Data:")
print(f"  Sample Mean (x̄) = Rs. {sample_mean}")
print(f"  Theoretical Mean (μ) = Rs. {theoretical_mean}")
print(f"  Standard Deviation (σ) = {theoretical_sigma}")
print(f"  Sample Size (n) = {n}")
```

```python
# Calculate standard error and test statistic
standard_error = theoretical_sigma / np.sqrt(n)
z_statistic = (sample_mean - theoretical_mean) / standard_error

print(f"\nCalculations:")
print(f"  Standard Error (SE) = σ/√n = {theoretical_sigma}/√{n} = {standard_error:.2f}")
print(f"  Test Statistic (z) = (x̄ - μ) / SE")
print(f"                     = ({sample_mean} - {theoretical_mean}) / {standard_error:.2f}")
print(f"                     = {z_statistic:.4f}")
```

## SECTION 4: CRITICAL VALUE AND P-VALUE

```python
# ---------------------------------------------------------
# Step 3: Determine Critical Value and P-value
# ---------------------------------------------------------
print("\n" + "="*60)
print("STEP 3: DETERMINE CRITICAL VALUE AND P-VALUE")
print("="*60)

# For right-tailed test at α = 0.05
z_critical = stats.norm.ppf(1 - alpha)

# P-value for right-tailed test
# Since z is negative, P(Z > z_statistic) will be very large (close to 1)
p_value = 1 - stats.norm.cdf(z_statistic)

print(f"\nSignificance Level (α) = {alpha}")
print(f"Critical Value (z_critical) = {z_critical:.4f}")
print(f"P-value = P(Z > {z_statistic:.4f}) = {p_value:.6f}")
```

## SECTION 5: DECISION RULE

```python
# ---------------------------------------------------------
# Step 4: Make a Decision
# ---------------------------------------------------------
print("\n" + "="*60)
print("STEP 4: DECISION RULE")
print("="*60)

print(f"\nDecision Rule for Right-Tailed Test:")
print(f"  - Reject H₀ if z > z_critical ({z_critical:.4f})")
print(f"  - Reject H₀ if p-value < α ({alpha})")

print(f"\nComparison:")
print(f"  Test Statistic (z) = {z_statistic:.4f}")
print(f"  Critical Value = {z_critical:.4f}")
print(f"  Is z > z_critical? {z_statistic} > {z_critical:.4f}? {z_statistic > z_critical}")
print(f"\n  P-value = {p_value:.6f}")
print(f"  Is p-value < α? {p_value:.6f} < {alpha}? {p_value < alpha}")

if z_statistic > z_critical:
    decision = "REJECT the Null Hypothesis (H₀)"
else:
    decision = "FAIL TO REJECT the Null Hypothesis (H₀)"

print(f"\n" + "*"*60)
print(f"DECISION: {decision}")
print("*"*60)
```

## SECTION 6: CONCLUSION

```python
# ---------------------------------------------------------
# Step 5: Conclusion with Clear Interpretation
# ---------------------------------------------------------
print("\n" + "="*60)
print("STEP 5: CONCLUSION")
print("="*60)

print(f"\nInterpretation of Results:")
print(f"  The test statistic z = {z_statistic:.4f} is NEGATIVE and falls")
print(f"  far to the LEFT of the theoretical mean.")
print(f"\n  This means the sample mean (Rs. {sample_mean}) is SIGNIFICANTLY")
print(f"  LOWER than the theoretical mean (Rs. {theoretical_mean}).")
print(f"\n  The p-value of {p_value:.6f} is much greater than α = {alpha},")
print(f"  indicating we cannot reject the null hypothesis.")

print(f"\n" + "-"*60)
print("BUSINESS CONCLUSION:")
print("-"*60)

if z_statistic > z_critical:
    conclusion = (
        "There IS strong evidence to support the restaurant owners' claim\n"
        "  that weekly operating costs are HIGHER than the model predicts."
    )
else:
    conclusion = (
        "There is NO evidence to support the restaurant owners' claim\n"
        "  that weekly operating costs are HIGHER than the model predicts.\n\n"
        "  In fact, the sample data shows that the actual weekly operating\n"
        "  costs (Rs. 3,050) are SIGNIFICANTLY LOWER than what the cost model\n"
        "  predicts (Rs. 4,000).\n\n"
        "  The difference of Rs. 950 is statistically significant and suggests\n"
        "  that either:\n"
        "  1. The restaurant owners' concerns are unfounded, OR\n"
        "  2. The sample data may not accurately represent their actual costs.\n\n"
        "  We FAIL TO REJECT the null hypothesis and conclude that costs\n"
        "  are NOT higher than the model suggests."
    )

print(f"\n{conclusion}")
print("\n" + "="*60 + "\n")
```

## SECTION 7: VISUALIZATION

```python
# ---------------------------------------------------------
# Visualization of the Hypothesis Test
# ---------------------------------------------------------

# Create the sampling distribution
x = np.linspace(theoretical_mean - 4*standard_error, theoretical_mean + 4*standard_error, 1000)
y = stats.norm.pdf(x, theoretical_mean, standard_error)

# Create the plot
plt.figure(figsize=(12, 7))
plt.plot(x, y, 'b-', linewidth=2, label='Sampling Distribution under H₀')

# Shade rejection region (right tail)
rejection_region_x = x[x >= theoretical_mean + z_critical*standard_error]
rejection_region_y = stats.norm.pdf(rejection_region_x, theoretical_mean, standard_error)
plt.fill_between(rejection_region_x, rejection_region_y,
                 color='red', alpha=0.3, label=f'Rejection Region (α={alpha})')

# Mark critical value
critical_value_x = theoretical_mean + z_critical * standard_error
plt.axvline(critical_value_x, color='red', linestyle='--', linewidth=2,
            label=f'Critical Value = Rs. {critical_value_x:.2f}')

# Mark theoretical mean
plt.axvline(theoretical_mean, color='blue', linestyle='-', linewidth=2,
            label=f'Theoretical Mean (μ) = Rs. {theoretical_mean}')

# Mark sample mean
plt.axvline(sample_mean, color='green', linestyle='--', linewidth=2.5,
            label=f'Sample Mean (x̄) = Rs. {sample_mean}')

# Add annotations
plt.annotate('Sample mean is FAR BELOW\nthe theoretical mean',
             xy=(sample_mean, stats.norm.pdf(sample_mean, theoretical_mean, standard_error)),
             xytext=(sample_mean - 200, 0.012),
             arrowprops=dict(arrowstyle='->', color='green', lw=2),
             fontsize=11, color='green', weight='bold',
             bbox=dict(boxstyle='round,pad=0.5', facecolor='lightgreen', alpha=0.7))

plt.annotate(f'z = {z_statistic:.2f}\n(highly negative)',
             xy=(sample_mean, 0.003),
             fontsize=10, color='darkgreen', weight='bold',
             ha='center')

plt.title('Hypothesis Test: Are Weekly Operating Costs Higher?\n(Right-Tailed Test)',
          fontsize=14, weight='bold')
plt.xlabel('Weekly Operating Cost (Rs.)', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, alpha=0.3)

# Add text box with conclusion
textstr = f'Decision: FAIL TO REJECT H₀\n\nConclusion: No evidence that costs\nare higher than model predicts.\n\nSample mean is Rs. 950 LOWER\nthan theoretical mean!'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
plt.text(0.02, 0.98, textstr, transform=plt.gca().transAxes, fontsize=10,
         verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("KEY OBSERVATION FROM VISUALIZATION:")
print("="*60)
print("\nThe green line (sample mean) is positioned far to the LEFT of")
print("the blue line (theoretical mean), NOT to the right.")
print("\nThe sample mean does not fall in the red rejection region.")
print("This visual confirms our conclusion: costs are LOWER, not HIGHER.")
print("="*60)
```

# Estimation and Confidence Intervals
## Print-Head Durability Analysis

### Background
In quality control processes, especially when dealing with high-value items, destructive sampling is a necessary but costly method to ensure product quality. The test to determine whether an item meets the quality standards destroys the item, leading to the requirement of small sample sizes due to cost constraints.

### Scenario
A manufacturer of print-heads for personal computers is interested in estimating the mean durability of their print-heads in terms of the number of characters printed before failure.

### Data
15 print-heads were randomly selected and tested until failure. Durability recorded in millions of characters.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd

# =============================================================================
# DATA INPUT
# =============================================================================
print("="*70)
print("CONFIDENCE INTERVAL ANALYSIS: PRINT-HEAD DURABILITY")
print("="*70)

# Durability data in millions of characters
durability = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
                       1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

print(f"\nData (Durability in millions of characters):")
print(durability)
print(f"\nSample Size (n) = {len(durability)}")

# Basic descriptive statistics
n = len(durability)
sample_mean = np.mean(durability)
sample_std = np.std(durability, ddof=1)  # Sample standard deviation (using n-1)
sample_variance = np.var(durability, ddof=1)

print(f"\n{'='*70}")
print("DESCRIPTIVE STATISTICS")
print("="*70)
print(f"Sample Mean (x̄)           = {sample_mean:.4f} million characters")
print(f"Sample Std Dev (s)        = {sample_std:.4f} million characters")
print(f"Sample Variance (s²)      = {sample_variance:.4f}")
print(f"Minimum                   = {np.min(durability):.4f} million characters")
print(f"Maximum                   = {np.max(durability):.4f} million characters")
print(f"Range                     = {np.max(durability) - np.min(durability):.4f} million characters")
print(f"Median                    = {np.median(durability):.4f} million characters")

# Create a summary table
print(f"\n{'='*70}")
print("DATA SUMMARY TABLE")
print("="*70)
df = pd.DataFrame({
    'Print-Head #': range(1, n+1),
    'Durability (millions)': durability
})
print(df.to_string(index=False))
print("="*70)
```

## Task A: 99% Confidence Interval Using Sample Standard Deviation

### Why Use t-Distribution?
When the population standard deviation (σ) is **unknown** and we use the sample standard deviation (s) as an estimate, we must account for the additional uncertainty. The t-distribution is wider than the normal distribution to reflect this uncertainty, especially for small sample sizes.

### Key Characteristics:
- Small sample size (n = 15)
- Population standard deviation is **unknown**
- Using sample standard deviation (s) as estimate
- Degrees of freedom = n - 1 = 14

### Formula:
$$CI = \bar{x} \pm t_{\alpha/2, df} \times \frac{s}{\sqrt{n}}$$

Where:
- $\bar{x}$ = sample mean
- $t_{\alpha/2, df}$ = critical t-value
- $s$ = sample standard deviation
- $n$ = sample size
- $df$ = degrees of freedom (n-1)

```python
# =============================================================================
# TASK A: 99% CONFIDENCE INTERVAL USING t-DISTRIBUTION
# =============================================================================

print("\n" + "="*70)
print("TASK A: 99% CONFIDENCE INTERVAL (Using Sample Standard Deviation)")
print("="*70)

# Parameters
confidence_level = 0.99
alpha = 1 - confidence_level
df = n - 1  # Degrees of freedom

print(f"\nStep 1: Identify Parameters")
print("-"*70)
print(f"Confidence Level           = {confidence_level*100}%")
print(f"Significance Level (α)     = {alpha}")
print(f"α/2                        = {alpha/2}")
print(f"Sample Size (n)            = {n}")
print(f"Degrees of Freedom (df)    = n - 1 = {df}")
print(f"Sample Mean (x̄)            = {sample_mean:.4f} million characters")
print(f"Sample Std Dev (s)         = {sample_std:.4f} million characters")

# Step 2: Find critical t-value
# For 99% CI, we need t-value that leaves 0.5% in each tail
t_critical = stats.t.ppf(1 - alpha/2, df)

print(f"\nStep 2: Find Critical t-value")
print("-"*70)
print(f"t-critical (t₀.₀₀₅, df=14) = {t_critical:.4f}")
print(f"\nNote: This value is from the t-distribution table with df=14")
print(f"      and α/2 = 0.005 (two-tailed test)")

# Step 3: Calculate Standard Error
standard_error_t = sample_std / np.sqrt(n)

print(f"\nStep 3: Calculate Standard Error")
print("-"*70)
print(f"Standard Error (SE) = s / √n")
print(f"                    = {sample_std:.4f} / √{n}")
print(f"                    = {sample_std:.4f} / {np.sqrt(n):.4f}")
print(f"                    = {standard_error_t:.4f} million characters")

# Step 4: Calculate Margin of Error
margin_error_t = t_critical * standard_error_t

print(f"\nStep 4: Calculate Margin of Error")
print("-"*70)
print(f"Margin of Error (ME) = t-critical × SE")
print(f"                     = {t_critical:.4f} × {standard_error_t:.4f}")
print(f"                     = {margin_error_t:.4f} million characters")

# Step 5: Calculate Confidence Interval
ci_lower_t = sample_mean - margin_error_t
ci_upper_t = sample_mean + margin_error_t

print(f"\nStep 5: Calculate 99% Confidence Interval")
print("-"*70)
print(f"Lower Limit = x̄ - ME = {sample_mean:.4f} - {margin_error_t:.4f} = {ci_lower_t:.4f}")
print(f"Upper Limit = x̄ + ME = {sample_mean:.4f} + {margin_error_t:.4f} = {ci_upper_t:.4f}")

print(f"\n{'*'*70}")
print(f"99% CONFIDENCE INTERVAL (t-distribution):")
print(f"[{ci_lower_t:.4f}, {ci_upper_t:.4f}] million characters")
print(f"{'*'*70}")

print(f"\nInterpretation:")
print("-"*70)
print(f"We are 99% confident that the true mean durability of all print-heads")
print(f"is between {ci_lower_t:.4f} and {ci_upper_t:.4f} million characters.")
print(f"\nThis means that if we repeated this sampling process many times and")
print(f"calculated a 99% CI each time, approximately 99% of those intervals")
print(f"would contain the true population mean.")

print(f"\nWhy t-Distribution?")
print("-"*70)
print(f"1. Small sample size (n = {n} < 30)")
print(f"2. Population standard deviation (σ) is UNKNOWN")
print(f"3. Using sample standard deviation (s) as an estimate")
print(f"4. t-distribution accounts for additional uncertainty from estimating σ")
print(f"5. As sample size increases, t-distribution approaches normal distribution")
```

## Task B: 99% Confidence Interval Using Known Population Standard Deviation

### Why Use z-Distribution?
When the population standard deviation (σ) is **known**, we use the normal (z) distribution instead of the t-distribution. This is because we don't need to estimate σ from the sample, eliminating one source of uncertainty.

### Key Characteristics:
- Population standard deviation is **known** (σ = 0.2 million characters)
- No need to estimate σ from sample
- Use z-distribution (standard normal distribution)
- Narrower confidence interval than t-distribution

### Formula:
$$CI = \bar{x} \pm z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}$$

Where:
- $\bar{x}$ = sample mean
- $z_{\alpha/2}$ = critical z-value
- $\sigma$ = **known** population standard deviation
- $n$ = sample size

```python
# =============================================================================
# TASK B: 99% CONFIDENCE INTERVAL USING z-DISTRIBUTION
# =============================================================================

print("\n" + "="*70)
print("TASK B: 99% CONFIDENCE INTERVAL (Using Known Population Std Dev)")
print("="*70)

# Known population standard deviation
population_std = 0.2

print(f"\nStep 1: Identify Parameters")
print("-"*70)
print(f"Confidence Level              = {confidence_level*100}%")
print(f"Significance Level (α)        = {alpha}")
print(f"α/2                           = {alpha/2}")
print(f"Sample Size (n)               = {n}")
print(f"Sample Mean (x̄)               = {sample_mean:.4f} million characters")
print(f"Population Std Dev (σ) KNOWN  = {population_std:.4f} million characters")

# Step 2: Find critical z-value
# For 99% CI, we need z-value that leaves 0.5% in each tail
z_critical = stats.norm.ppf(1 - alpha/2)

print(f"\nStep 2: Find Critical z-value")
print("-"*70)
print(f"z-critical (z₀.₀₀₅)           = {z_critical:.4f}")
print(f"\nNote: This value is from the standard normal (z) distribution table")
print(f"      with α/2 = 0.005 (two-tailed test)")
print(f"      Standard z-value for 99% CI is approximately 2.576")

# Step 3: Calculate Standard Error
standard_error_z = population_std / np.sqrt(n)

print(f"\nStep 3: Calculate Standard Error")
print("-"*70)
print(f"Standard Error (SE) = σ / √n")
print(f"                    = {population_std:.4f} / √{n}")
print(f"                    = {population_std:.4f} / {np.sqrt(n):.4f}")
print(f"                    = {standard_error_z:.4f} million characters")

# Step 4: Calculate Margin of Error
margin_error_z = z_critical * standard_error_z

print(f"\nStep 4: Calculate Margin of Error")
print("-"*70)
print(f"Margin of Error (ME) = z-critical × SE")
print(f"                     = {z_critical:.4f} × {standard_error_z:.4f}")
print(f"                     = {margin_error_z:.4f} million characters")

# Step 5: Calculate Confidence Interval
ci_lower_z = sample_mean - margin_error_z
ci_upper_z = sample_mean + margin_error_z

print(f"\nStep 5: Calculate 99% Confidence Interval")
print("-"*70)
print(f"Lower Limit = x̄ - ME = {sample_mean:.4f} - {margin_error_z:.4f} = {ci_lower_z:.4f}")
print(f"Upper Limit = x̄ + ME = {sample_mean:.4f} + {margin_error_z:.4f} = {ci_upper_z:.4f}")

print(f"\n{'*'*70}")
print(f"99% CONFIDENCE INTERVAL (z-distribution):")
print(f"[{ci_lower_z:.4f}, {ci_upper_z:.4f}] million characters")
print(f"{'*'*70}")

print(f"\nInterpretation:")
print("-"*70)
print(f"We are 99% confident that the true mean durability of all print-heads")
print(f"is between {ci_lower_z:.4f} and {ci_upper_z:.4f} million characters.")
print(f"\nThis interval is based on the KNOWN population standard deviation of")
print(f"{population_std} million characters.")

print(f"\nWhy z-Distribution?")
print("-"*70)
print(f"1. Population standard deviation (σ = {population_std}) is KNOWN")
print(f"2. No need to estimate σ from the sample")
print(f"3. Less uncertainty compared to using sample standard deviation")
print(f"4. z-distribution (normal distribution) is appropriate when σ is known")
print(f"5. Results in a narrower confidence interval (more precise estimate)")
```

```python
# =============================================================================
# VISUALIZATION: COMPARISON OF CONFIDENCE INTERVALS
# =============================================================================

fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# -------------------------------------------------------------------------
# Plot 1: Confidence Intervals Comparison
# -------------------------------------------------------------------------
ax1 = axes[0]

# Plot both intervals
y_positions = [2, 1]
intervals = [(ci_lower_t, ci_upper_t), (ci_lower_z, ci_upper_z)]
colors = ['blue', 'green']
labels = ['99% CI (t-distribution)', '99% CI (z-distribution)']

for i, (interval, color, label) in enumerate(zip(intervals, colors, labels)):
    lower, upper = interval
    y = y_positions[i]

    # Plot the interval line
    ax1.plot([lower, upper], [y, y], color=color, linewidth=3, label=label)

    # Plot the endpoints
    ax1.plot([lower, upper], [y, y], 'o', color=color, markersize=8)

    # Plot the center (sample mean)
    ax1.plot(sample_mean, y, 'D', color=color, markersize=10)

    # Add interval annotations
    ax1.text(lower, y - 0.15, f'{lower:.4f}', ha='center', fontsize=9, color=color, weight='bold')
    ax1.text(upper, y - 0.15, f'{upper:.4f}', ha='center', fontsize=9, color=color, weight='bold')

# Add vertical line at sample mean
ax1.axvline(sample_mean, color='red', linestyle='--', linewidth=2,
            label=f'Sample Mean = {sample_mean:.4f}', alpha=0.7)

# Formatting
ax1.set_yticks(y_positions)
ax1.set_yticklabels(['Task A\n(t-dist)', 'Task B\n(z-dist)'], fontsize=11)
ax1.set_xlabel('Durability (millions of characters)', fontsize=12, weight='bold')
ax1.set_title('Comparison of 99% Confidence Intervals', fontsize=14, weight='bold', pad=20)
ax1.legend(loc='upper right', fontsize=10)
ax1.grid(axis='x', alpha=0.3, linestyle='--')
ax1.set_ylim(0.5, 2.5)

# Add annotation box
textstr = f'Note: t-interval is WIDER\ndue to additional uncertainty\nfrom estimating σ'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
ax1.text(0.02, 0.98, textstr, transform=ax1.transAxes, fontsize=10,
         verticalalignment='top', bbox=props)

# -------------------------------------------------------------------------
# Plot 2: Data Distribution with Confidence Intervals
# -------------------------------------------------------------------------
ax2 = axes[1]

# Create histogram of data
ax2.hist(durability, bins=8, alpha=0.6, color='skyblue', edgecolor='black',
         label='Sample Data Distribution')

# Add vertical lines for mean and CIs
ax2.axvline(sample_mean, color='red', linewidth=2.5,
            label=f'Sample Mean = {sample_mean:.4f}', linestyle='-')

# t-distribution CI
ax2.axvline(ci_lower_t, color='blue', linewidth=2, linestyle='--',
            label=f't-CI: [{ci_lower_t:.4f}, {ci_upper_t:.4f}]')
ax2.axvline(ci_upper_t, color='blue', linewidth=2, linestyle='--')
ax2.axvspan(ci_lower_t, ci_upper_t, alpha=0.2, color='blue')

# z-distribution CI
ax2.axvline(ci_lower_z, color='green', linewidth=2, linestyle='-.',
            label=f'z-CI: [{ci_lower_z:.4f}, {ci_upper_z:.4f}]')
ax2.axvline(ci_upper_z, color='green', linewidth=2, linestyle='-.')
ax2.axvspan(ci_lower_z, ci_upper_z, alpha=0.2, color='green')

# Formatting
ax2.set_xlabel('Durability (millions of characters)', fontsize=12, weight='bold')
ax2.set_ylabel('Frequency', fontsize=12, weight='bold')
ax2.set_title('Sample Data Distribution with 99% Confidence Intervals',
              fontsize=14, weight='bold', pad=20)
ax2.legend(loc='upper right', fontsize=9)
ax2.grid(axis='y', alpha=0.3, linestyle='--')

# Add summary statistics box
stats_text = f'Summary Statistics:\n'
stats_text += f'n = {n}\n'
stats_text += f'x̄ = {sample_mean:.4f}\n'
stats_text += f's = {sample_std:.4f}\n'
stats_text += f'σ (known) = {population_std:.4f}'
props2 = dict(boxstyle='round', facecolor='lightblue', alpha=0.8)
ax2.text(0.02, 0.98, stats_text, transform=ax2.transAxes, fontsize=10,
         verticalalignment='top', bbox=props2, family='monospace')

plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("VISUALIZATION INSIGHTS")
print("="*70)
print("\nTop Panel: Directly compares the two confidence intervals")
print("  - Blue interval (t-dist) is wider due to estimation uncertainty")
print("  - Green interval (z-dist) is narrower due to known σ")
print("  - Both are centered at the same sample mean")

print("\nBottom Panel: Shows how CIs relate to the actual data")
print("  - Histogram shows the distribution of our 15 sample points")
print("  - Blue shaded region: 99% CI using t-distribution")
print("  - Green shaded region: 99% CI using z-distribution")
print("  - Red line: Sample mean (center of both intervals)")
print("="*70)
```
