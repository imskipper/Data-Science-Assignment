```python
def check_number(number):
    if number > 0:
        if number % 2 == 0:
            print(f"{number} is positive and even.")
        else:
            print(f"{number} is positive but odd.")
    else:
        if number == 0:
            print(f"{number} is zero.")
        else:
            print(f"{number} is negative.")

# Example usage:
check_number(4)
check_number(7)
check_number(0)
check_number(-3)
```

**Section C: Exploring Pandas**

11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
import numpy as np

# Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Alice', 'Frank', 'Grace', 'Heidi', 'Ivan'],
    'Age': [25, 30, 35, 28, 22, 25, 40, 32, 29, 38],
    'City': ['New York', 'London', 'Paris', 'New York', 'Berlin', 'New York', 'Rome', 'London', 'Paris', 'Berlin'],
    'Salary': [50000, 60000, 75000, np.nan, 55000, 50000, 80000, 62000, np.nan, 70000]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\n")

# Use .info()
print("DataFrame Info:")
df.info()
print("\n")

# Use .describe()
print("DataFrame Description:")
print(df.describe())
print("\n")

# Select 'Name' and 'City' columns
print("Selected 'Name' and 'City' columns:")
print(df[['Name', 'City']])
print("\n")

# Drop 'City' column
df_no_city = df.drop('City', axis=1)
print("DataFrame after dropping 'City' column:")
print(df_no_city)
print("\n")

# Fill any missing values in 'Salary' column with the mean
mean_salary = df['Salary'].mean()
df['Salary'].fillna(mean_salary, inplace=True)
print("DataFrame after filling missing 'Salary' values with the mean:")
print(df)
print("\n")

# Remove any duplicate rows
df_duplicates = df.drop_duplicates()
print("DataFrame after removing duplicate rows:")
print(df_duplicates)
```
